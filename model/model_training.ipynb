```python
import pandas as pd
import numpy as np
import io
import re

# Load CSV
df = pd.read_csv(io.StringIO(csv_content), engine="python")

# Clean harga
def clean_price(price):
    if pd.isna(price):
        return np.nan
    s = str(price).lower().replace("rp", "").replace(" ", "")
    m = re.search(r"([\d,.]+)(jt|m)", s)
    if not m:
        return np.nan
    value = float(m.group(1).replace(",", "."))
    return value * (1_000_000 if m.group(2) == "jt" else 1_000_000_000)

df["Harga (IDR)"] = df["harga"].apply(clean_price)

# Clean daerah
df["subdistrict"] = df["daerah"].str.replace(",", "").str.strip()

# Numerik
df["Luas Tanah (m²)"] = pd.to_numeric(df["luas_tanah_m²"], errors="coerce")
df["Luas Bangunan (m²)"] = pd.to_numeric(df["luas_bangunan_m²"], errors="coerce")
df["Kamar Tidur"] = pd.to_numeric(df["kamar_tidur"], errors="coerce")
df["Kamar Mandi"] = pd.to_numeric(df["kamar_mandi"], errors="coerce")

# Filter NaN
df = df.dropna(subset=[
    "Harga (IDR)",
    "Luas Tanah (m²)",
    "Luas Bangunan (m²)",
    "Kamar Tidur",
    "Kamar Mandi",
    "subdistrict"
])

# One-hot encoding
df_dummies = pd.get_dummies(
    df["subdistrict"],
    prefix="Daerah",
    drop_first=True,
    dtype=int
)

# Final dataset
model_df = pd.concat([
    df[["Luas Tanah (m²)", "Luas Bangunan (m²)", "Kamar Tidur", "Kamar Mandi"]],
    df_dummies,
    df["Harga (IDR)"]
], axis=1)

print("model_df siap dipakai:", model_df.shape)
display(model_df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# =========================================================
# 1. DESCRIPTIVE STATISTICS
# =========================================================
pd.set_option("display.float_format", lambda x: f"{x:,.2f}")

print("=== Statistik Awal Dataset ===")
display(model_df.describe())

# =========================================================
# 2. DETEKSI OUTLIER MENGGUNAKAN METODE IQR
# =========================================================
fitur_outlier = [
    'Luas Tanah (m²)',
    'Luas Bangunan (m²)',
    'Kamar Tidur',
    'Kamar Mandi',
    'Harga (IDR)'
]

all_outliers = []

for col in fitur_outlier:
    Q1 = model_df[col].quantile(0.25)
    Q3 = model_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    mask = (model_df[col] < lower) | (model_df[col] > upper)
    outlier_rows = model_df.loc[mask].copy()
    outlier_rows["Outlier_Dari"] = col

    all_outliers.append(outlier_rows)

# Gabungkan dan hilangkan duplikat
outliers_combined = pd.concat(all_outliers)
outliers_combined = outliers_combined.loc[
    ~outliers_combined.index.duplicated(keep="first")
]

print(f"Jumlah outlier unik terdeteksi (IQR): {len(outliers_combined)}")
display(outliers_combined)

# =========================================================
# 3. DATASET TANPA OUTLIER (IQR)
# =========================================================
df_clean_iqr = model_df.drop(index=outliers_combined.index).copy()

print("\n=== Ringkasan Cleaning (IQR) ===")
print(f"Data awal           : {len(model_df)}")
print(f"Setelah outlier drop: {len(df_clean_iqr)}")
print(f"Data dihapus        : {len(outliers_combined)}")

display(df_clean_iqr.describe())
```

```python
def plot_korelasi(x, y, xlabel, ylabel, title):
    plt.figure(figsize=(8,5))
    plt.scatter(x, y, marker='+')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(alpha=0.3)
    plt.show()

    corr = pd.DataFrame({xlabel: x, ylabel: y}).corr().iloc[0,1]
    print(f"Korelasi {xlabel} vs {ylabel} = {corr:.3f}")
    print("-"*60)

# Kamar Tidur
plot_korelasi(
    df_clean_iqr['Kamar Tidur'],
    df_clean_iqr['Harga (IDR)'],
    'Kamar Tidur',
    'Harga (IDR)',
    'Korelasi Jumlah Kamar Tidur dengan Harga Rumah'
)

# Kamar Mandi
plot_korelasi(
    df_clean_iqr['Kamar Mandi'],
    df_clean_iqr['Harga (IDR)'],
    'Kamar Mandi',
    'Harga (IDR)',
    'Korelasi Jumlah Kamar Mandi dengan Harga Rumah'
)

# Luas Tanah
plot_korelasi(
    df_clean_iqr['Luas Tanah (m²)'],
    df_clean_iqr['Harga (IDR)'],
    'Luas Tanah (m²)',
    'Harga (IDR)',
    'Korelasi Luas Tanah dengan Harga Rumah'
)

# Luas Bangunan
plot_korelasi(
    df_clean_iqr['Luas Bangunan (m²)'],
    df_clean_iqr['Harga (IDR)'],
    'Luas Bangunan (m²)',
    'Harga (IDR)',
    'Korelasi Luas Bangunan dengan Harga Rumah'
)
```
